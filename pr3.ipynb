```python
null_col=[]

for col in data.columns:
    if data[col].isnull().any()==True:
        print(f"{col}")
        null_col.append(col)
        
        
```

```python
for col in null_col:
    if data[col].isnull().sum()==1:
        data=data.dropna(subset=[col])
    elif data[col].dtype=='float' or data[col].dtype=='int':
        data=data.fillna(data[col].mean())
    else:
        data=data.fillna(data[col].mode()[0])        
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['Salary'] = pd.to_numeric(data['Salary'], errors='coerce')

```

```python
for col in data.columns:
    if data[col].dtype =='float' or data[col].dtype =='int':
        print(f"{col} : {data[col].mean()}")
```

```python
for col in data.columns:
    if data[col].dtype =='float' or data[col].dtype =='int':
        print(f"{col} : {np.mean(data[col])}")
```

```python
for col in data.columns:
    if data[col].dtype =='float' or data[col].dtype =='int':
        print(f"{col} : {data[col].median()}")
```

```python
for col in data.columns:
    if data[col].dtype =='float' or data[col].dtype =='int':
        print(f"{col} : {np.median(data[col])}")
```

```python
for col in data.columns:
    print(f"{col}:{data[col].mode()[0]}")
```

```python
for col in data.columns:
    if data[col].dtype =='float' or data[col].dtype =='int':
        print(f"{col} : {data[col].min()}")

for col in data.columns:
    if data[col].dtype =='float' or data[col].dtype =='int':
        print(f"{col} : {np.min(data[col])}")      
```

```python
for col in data.columns:
    if data[col].dtype =='float' or data[col].dtype =='int':
        print(f"{col} : {data[col].max()}")

for col in data.columns:
    if data[col].dtype =='float' or data[col].dtype =='int':
        print(f"{col} : {np.max(data[col])}")        
```

```python
for col in data.columns:
    if data[col].dtype =='float' or data[col].dtype =='int':
        print(f"{col} : {data[col].std()}")

for col in data.columns:
    if data[col].dtype =='float' or data[col].dtype =='int':
        print(f"{col} : {np.std(data[col])}")  
```

```python
grouped_data=data.groupby('Age')['Salary'].describe()
grouped_data
```

```python
grouped_data=data.groupby('Age')['Salary'].agg(
     Mean='mean',
    Median='median',
    Minimum='min',
    Maximum='max',
    Std_Dev='std')
grouped_data
```

```python
# right=False means if 19-23 the take 19 dont take 23
data['AgeGroup'] = pd.cut(data['Age'], bins=[19, 23, 27, 31, 35,39,43], labels=['19-22', '23-26', '27-30', '31-34','35-38','39-42'],right=False)
grouped_data=data.groupby('AgeGroup')['Salary'].describe()
grouped_data
```
