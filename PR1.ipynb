```python
#replace the null value of salary with median
data['Salary'].fillna(data['Salary'].median())
data
```

```python
#data normalization and formating

#convert the position from object to category

data['Position']=data['Position'].astype('category')
data
```

```python
data.dtypes
```

```python
# Create a new column 'Position_Code' that maps the positions to integer codes
data['Position_code']=data['Position'].cat.codes
# Display the updated DataFrame
print(data[['Name', 'Position', 'Position_code']])
```

```python
# convert object to category in loop
categorical_cols=['Team','College','Height'] 

for col in categorical_cols:
    data[col]=data[col].astype('category')
```

```python
#One hot encoding
df_encoded = pd.get_dummies(data, columns=['Height','Team','College'])
df_encoded.head()
```
